```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
```

```python
# Manual solution
```

```python
frame = cv2.imread('./frames/frame_19.jpg')

# Convert frame to grayscale
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Convert the frame to grayscale
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```

```python
# Define the Sobel kernels for edge detection
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])
```

```python
# Apply Sobel operators for gradient calculation
gradient_x = cv2.filter2D(gray, -1, sobel_x)
gradient_y = cv2.filter2D(gray, -1, sobel_y)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gradient_x, cmap='gray')
plt.title('Gradient X')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradient_y, cmap='gray')
plt.title('Gradient Y')
plt.axis('off')

plt.show()
```

```python
# Calculate magnitude and direction of gradients
magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
angle = np.arctan2(gradient_y, gradient_x)
```

```python
# Non-maximum suppression
edge_magnitude = np.zeros_like(gray)
for i in range(1, gray.shape[0]-1):
    for j in range(1, gray.shape[1]-1):
        direction = angle[i, j]
        if (direction >= -np.pi/8 and direction < np.pi/8) or (direction >= 7*np.pi/8 or direction < -7*np.pi/8):
            if (magnitude[i, j] >= magnitude[i, j+1]) and (magnitude[i, j] >= magnitude[i, j-1]):
                edge_magnitude[i, j] = magnitude[i, j]
        elif (direction >= np.pi/8 and direction < 3*np.pi/8) or (direction >= -7*np.pi/8 and direction < -5*np.pi/8):
            if (magnitude[i, j] >= magnitude[i-1, j+1]) and (magnitude[i, j] >= magnitude[i+1, j-1]):
                edge_magnitude[i, j] = magnitude[i, j]
        elif (direction >= 3*np.pi/8 and direction < 5*np.pi/8) or (direction >= -5*np.pi/8 and direction < -3*np.pi/8):
            if (magnitude[i, j] >= magnitude[i-1, j]) and (magnitude[i, j] >= magnitude[i+1, j]):
                edge_magnitude[i, j] = magnitude[i, j]
        else:
            if (magnitude[i, j] >= magnitude[i-1, j-1]) and (magnitude[i, j] >= magnitude[i+1, j+1]):
                edge_magnitude[i, j] = magnitude[i, j]
```

```python
# Double thresholding
threshold_low = 10
threshold_high = 40
edge_magnitude[edge_magnitude < threshold_low] = 0
edge_magnitude[edge_magnitude > threshold_high] = 255
```

```python
# Display the manually detected edges
plt.imshow(edge_magnitude, cmap='gray')
plt.title('Canny Edge Detection (Manual)')
plt.show()
```

```python
# Open CV
frame = cv2.imread('./frames/frame_19.jpg')

# Convert frame to grayscale
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Apply Canny Edge Detection using OpenCV
edges_opencv = cv2.Canny(gray, 50, 150)
```

```python
# Display the OpenCV Canny Edge Detection result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale Image')
plt.subplot(1, 2, 2)
plt.imshow(edges_opencv, cmap='gray')
plt.title('Canny Edge Detection (OpenCV)')
plt.show()
```

```python
# Comparison of the two solutions
# The manual implementation was able to detect edges that are vertices of the object as well as those that are edges on the illustration on the object.

# The OpenCV implementation was able to detect everything mentioned above, but at higher accuracy, even capturing the edges on the font.

# The OpenCV implementation was more accurate.
```
