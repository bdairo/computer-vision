```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
```

```python
# Manual solution
frame = cv2.imread('./frames/frame_19.jpg')

# Convert frame to grayscale
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Convert the frame to grayscale
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```

```python
# Define the size of the patch
patch_size = 20
half_patch = patch_size // 2
```

```python
# Select the center of the frame as the patch
center_x = width // 2
center_y = height // 2
patch_manual = gray[center_y - half_patch:center_y + half_patch + 1,
                    center_x - half_patch:center_x + half_patch + 1]
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(patch_manual, cmap='gray')
plt.title('Selected Patch for Manual Detection')
```

```python
# Manually perform Harris Corner Detection on the patch
# Compute gradients using Sobel operator
dx = cv2.Sobel(patch_manual, cv2.CV_64F, 1, 0, ksize=3)
dy = cv2.Sobel(patch_manual, cv2.CV_64F, 0, 1, ksize=3)
```

```python
# Compute products of gradients
dx2 = dx ** 2
dy2 = dy ** 2
dxy = dx * dy
```

```python
# Compute sums of products in a local window
window_size = 3
offset = window_size // 2
corner_response = np.zeros_like(patch_manual, dtype=np.float32)
```

```python
for y in range(offset, patch_size - offset):
    for x in range(offset, patch_size - offset):
        sum_dx2 = np.sum(dx2[y - offset:y + offset + 1, x - offset:x + offset + 1])
        sum_dy2 = np.sum(dy2[y - offset:y + offset + 1, x - offset:x + offset + 1])
        sum_dxy = np.sum(dxy[y - offset:y + offset + 1, x - offset:x + offset + 1])

        # Compute the determinant and trace of the matrix
        det = (sum_dx2 * sum_dy2) - (sum_dxy ** 2)
        trace = sum_dx2 + sum_dy2

        # Compute the Harris corner response
        corner_response[y, x] = det - 0.04 * (trace ** 2)
```

```python
# Threshold the corner response to find corners
threshold = 0.01 * np.max(corner_response)
corners_manual = np.zeros_like(patch_manual)
corners_manual[corner_response > threshold] = 255
```

```python
# Display the Manual Harris Corner Detection result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(patch_manual, cmap='gray')
plt.title('Selected Patch for Manual Detection')
plt.subplot(1, 2, 2)
plt.imshow(corners_manual, cmap='gray')
plt.title('Harris Corner Detection (Manual)')
plt.show()
```

```python
# Open CV Implementation
opencv_corners = cv2.cornerHarris(patch_manual, blockSize=2, ksize=3, k=0.04)
opencv_corners = cv2.dilate(opencv_corners, None)
```
